```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

def apply_average_filter(image, kernel_size):
    # Apply the average filter with the specified kernel size
    return cv2.blur(image, (kernel_size, kernel_size))

def calculate_psnr(original_image, filtered_image):
    # Calculate the Mean Squared Error (MSE)
    mse = np.mean((original_image - filtered_image) ** 2)

    # Calculate the maximum pixel value
    max_pixel = 255.0

    # Calculate the PSNR
    if mse == 0:
        return 100
    psnr = 20 * np.log10(max_pixel / np.sqrt(mse))
    return psnr

# Read the image
image_path = "/content/sample_data/Comment tailler les rosiers pour une nouvelle floraison _.jpg"
image = cv2.imread(image_path)

# Check if the image is loaded successfully
if image is None:
    print("Error: Unable to load the image.")
else:
    # Convert the image to grayscale
    grayscale_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Apply average filters with different kernel sizes
    kernel_sizes = [3, 5, 7, 9, 11, 13, 15]
    psnr_values = []
    for kernel_size in kernel_sizes:
        filtered_image = apply_average_filter(grayscale_image, kernel_size)

        # Calculate PSNR
        psnr = calculate_psnr(grayscale_image, filtered_image)
        psnr_values.append(psnr)

        # Display the filtered grayscale image
        plt.figure(figsize=(8, 6))
        plt.imshow(filtered_image, cmap='gray')
        plt.title(f"{kernel_size}x{kernel_size} Average Filter")
        plt.axis('off')
        plt.show()

    # Display PSNR values in a separate table
    plt.figure(figsize=(8, 6))
    table_data = [["Kernel Size", "PSNR (dB)"]]
    for i in range(len(kernel_sizes)):
        table_data.append([kernel_sizes[i], f"{psnr_values[i]:.2f}"])
    plt.table(cellText=table_data,
              cellLoc='center',
              loc='center')
    plt.axis('off')
    plt.show()

```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

def apply_laplacian(image, laplacian_mask):
    # Convert the image to grayscale
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Apply Laplacian filter
    laplacian = cv2.filter2D(gray_image, -1, laplacian_mask)

    # Sharpen the image using Laplacian filter
    sharpened_image = gray_image - laplacian

    return laplacian, sharpened_image

# Get Laplacian mask from user input
mask_values = input("Enter Laplacian mask values separated by spaces: ").strip().split()
laplacian_mask = np.array(mask_values, dtype=np.float32).reshape(3, 3)

# Read the image
image = cv2.imread('/content/Comment tailler les rosiers pour une nouvelle floraison _.jpg')

# Apply Laplacian filter and sharpen the image
laplacian_map, sharpened_image = apply_laplacian(image, laplacian_mask)

# Display the results
plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
plt.title('Original Image')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.subplot(1, 3, 2)
plt.title('Laplacian Map')
plt.imshow(np.abs(laplacian_map), cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.title('Sharpened Image')
plt.imshow(sharpened_image, cmap='gray')
plt.axis('off')

plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

def apply_gaussian_filter(image, kernel_size):
    # Apply the Gaussian filter with the specified kernel size
    return cv2.GaussianBlur(image, (kernel_size, kernel_size), 0)

def calculate_psnr(original_image, filtered_image):
    # Calculate the Mean Squared Error (MSE)
    mse = np.mean((original_image - filtered_image) ** 2)

    # Calculate the maximum pixel value
    max_pixel = 255.0

    # Calculate the PSNR
    if mse == 0:
        return 100
    psnr = 20 * np.log10(max_pixel / np.sqrt(mse))
    return psnr

# Read the image
image_path = "/content/sample_data/Comment tailler les rosiers pour une nouvelle floraison _.jpg"
image = cv2.imread(image_path)

# Check if the image is loaded successfully
if image is None:
    print("Error: Unable to load the image.")
else:
    # Convert the image to grayscale
    grayscale_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Apply Gaussian filters with different kernel sizes
    kernel_sizes = [3, 5, 7, 9, 11, 13, 15]
    psnr_values = []
    for kernel_size in kernel_sizes:
        filtered_image = apply_gaussian_filter(grayscale_image, kernel_size)

        # Calculate PSNR
        psnr = calculate_psnr(grayscale_image, filtered_image)
        psnr_values.append(psnr)

        # Display the filtered grayscale image
        plt.figure(figsize=(8, 6))
        plt.imshow(filtered_image, cmap='gray')
        plt.title(f"{kernel_size}x{kernel_size} Gaussian Filter")
        plt.axis('off')
        plt.show()

    # Display PSNR values in a separate table
    plt.figure(figsize=(8, 6))
    table_data = [["Kernel Size", "PSNR (dB)"]]
    for i in range(len(kernel_sizes)):
        table_data.append([kernel_sizes[i], f"{psnr_values[i]:.2f}"])
    plt.table(cellText=table_data,
              cellLoc='center',
              loc='center')
    plt.axis('off')
    plt.show()
```
